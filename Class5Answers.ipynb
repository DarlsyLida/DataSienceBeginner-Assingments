### Assignment 3

With pandas groupby method, count all states that belongs to each region

```python
#Creating those a function that can map states to different regions, we have:
def region(state):
    NC = ['Benue State', 'Kogi', 'Kwara', 'Nasarawa', 'Niger', 'Plateau', 'FCT']
    SS = ['Akwa', 'Bayelsa', 'Cross River', 'Rivers', 'Delta', 'Edo']
    SE = ['Abia', 'Anambra', 'Ebonyi', 'Enugu', 'Imo']
    SW = ['Ekiti', 'Lagos', 'Ogun', 'Ondo', 'Osun', 'Oyo']
    NE = ['Adamawa', 'Bauchi', 'Borno', 'Gombe', 'Taraba', 'Yobe']
    NW = ['Jigawa', 'Kaduna', 'Kano', 'Katsina', 'Kebbi', 'Sokoto', 'Zamfara']
    if state in NC:
        return 'North Central'
    elif state in SS:
        return 'South South'
    elif state in SE:
        return 'South East'
    elif state in SW:
        return 'South West'
    elif state in NE:
        return 'North East'
    else:
        return 'North West'
```

```python
pop_df['STATE'] = [state.split()[0] if len(state.split())<=2 else (state.split()[0] + ' ' + state.split()[1]) for state in (pop_df['STATES'])]

pop_df.head()
```

```python
pop_df.drop('STATES', axis=1, inplace=True)
```

```python
#Applying th efunction using pandas.apply() method, we have:

pop_df['Region'] = pop_df['STATE'].apply(region)

#Lets see whats in the new feature and the entire dataset
pop_df.head()
```

```python
temp = pop_df.groupby(['Region'])
temp.describe(include='all')
```
