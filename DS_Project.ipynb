#### - Question 5

We want to find out the most popular music genre for each country.we determine the most popular genre as the highest amount of purchases.Write a query that returns each country along with the top genre with the highest amount of purchases.Write a query that returns each country along with the top genre for countries where the maximum number of purchases is shared all genres.For this query,you will need to use the Invoice,InvoiceLine,Customer,Track,and Genre tables.

```python
invoice=pd.read_csv('Invoice.csv')
invoiceLine=pd.read_csv('InvoiceLine.csv')
customer=pd.read_csv('Customer.csv')
track=pd.read_csv('Track.csv')
genre=pd.read_csv('Genre.csv')
```

```python
invoice.info()
```

```python
invoice.drop(['CustomerId'],axis=1, inplace=True)
```

```python
invoiceLine.info()
```

```python
invoiceLine.drop(['InvoiceId'], axis=1, inplace=True)
```

```python
customer.info()
```

```python
track.info()
```

```python
track.drop(['GenreId'],axis=1, inplace=True)
```

```python
genre.info()
```

```python
data=[invoice, invoiceLine, customer, track, genre]
data2=pd.concat(data, axis=1)
data2
```

```python
data2.columns
```

```python
data2.info()
```

```python
topCountry=data2[['BillingCountry', 'GenreId']].groupby('BillingCountry').count()\
.sort_values(by='GenreId', ascending=False).round(2)
topCountry.rename(columns={'GenreId':'GenresPerCountry'},inplace=True)
print('The Top Country is:')
topCountry.head(1)
```

```python
print("Total Genres Per Country:")
topCountry
```
